```python
# importing orders_products_merged
df_ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
df = df_ords_prods_merge[:1000000]
```

```python
# categorizing prices
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else: return 'Not enough data'
```

```python
# create new column
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# count price categories
df['price_range'].value_counts(dropna = False)
```

```python
# count values of orders_day_of_week
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# determine busiest day
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value == 0:
        result.append("Busiest day")
    elif value == 4: 
        result.append("Least busy")
    else:
        result.append("Regularly busy")
```

```python
# check output
result
```

```python
# create new column on ords_prods_merge
df_ords_prods_merge['busiest_day'] = result
```

```python
# count values of busiest_day
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

```python
# create list of busiest_days
results = []

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value == 0 or value == 1:
        results.append("Busiest Days")
    elif value == 3 or value == 4:
        results.append("Slowest Days")
    else: 
        results.append("Regularly Busy Days")
```

```python
# create busiest_days column on ords_prods_merge
df_ords_prods_merge['busiest_days'] = results
```

```python
# count values of busiest_days
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

Counts for busiest_days compared to busiest_day are now much closer to each other as we are now comparing days in groups of 2,2, and 3 as opposed to 1,1, and 5. We confirmed above that the counts remain the same.

```python
# determine busiest hours of the day
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# create list of busiest hours
sorts = []

for value in df_ords_prods_merge["order_hour_of_day"]:
    if value in range(9,16):
        sorts.append("Most Orders")
    elif value in range(0,8):
        sorts.append("Fewest Orders")
    else:
        sorts.append("Average Orders")
```

```python
# check output 
sorts
```

```python
# create busiest_period_of_day column
df_ords_prods_merge['busiest_period_of_day'] = sorts
```

```python
# check frequency of busiest_period_of_day
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
